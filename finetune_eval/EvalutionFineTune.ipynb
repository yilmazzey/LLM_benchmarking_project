```python
import json
import sys
import os

# Add UniEval to Python path
sys.path.append(os.path.abspath("UniEval"))

from UniEval.utils import convert_to_json
from UniEval.metric.evaluator import get_evaluator

def load_abstracts_from_file(file_path):
    """Load abstracts from a JSON file.

    Args:
        file_path (str): Path to the JSON file.

    Returns:
        list: List of abstracts loaded from the file, or None if file not found or JSON error.
    """
    try:
        with open(file_path, 'r') as f:
            data = json.load(f)
        return data
    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")
        return None  # Return None to indicate failure
    except json.JSONDecodeError:
        print(f"Error: Invalid JSON format in {file_path}")
        return None  # Return None to indicate failure


def extract_abstracts(original_data, generated_data):
    """Extract original abstracts, generated abstracts, and source articles from loaded data.

    Args:
        original_data (list): List of dictionaries loaded from the original articles file.
        generated_data (list): List of dictionaries loaded from the generated abstracts file.

    Returns:
        tuple: A tuple containing three lists: src_list (source articles), output_list (generated abstracts),
               and ref_list (original abstracts). Returns empty lists if there are issues with input data.
    """
    src_list = []  # Source articles (full articles)
    output_list = [] # Generated abstracts
    ref_list = []   # Original abstracts

    if original_data is None or generated_data is None:
        print("Error: One or both input data lists are None. Check if files were loaded correctly.")
        return [], [], []

    if not isinstance(original_data, list) or not isinstance(generated_data, list):
        print("Error: Input data is not in the expected list format.")
        return [], [], []

    # Assuming both files have the same number of entries and are in the same order
    num_articles = min(len(original_data), len(generated_data)) # Handle cases with different lengths

    for i in range(num_articles):
        original_article = original_data[i]
        generated_article = generated_data[i]

        original_abstract = original_article['abstract'] #Get methodu kaldırıldı KeyError önemli
        generated_abstract = generated_article['generated_abstract']#Get methodu kaldırıldı KeyError önemli
        article_title = original_article['title'] # Get article title
        article_sections = original_article['content'] #yeni veride section yok content var ve direct access olarak değiştirildi
        source_article = ""
        print(source_article)
        print(generated_abstract)
        if article_title:
            source_article += article_title + "\n\n" # Add title and separator
        if article_sections:
            source_article += article_sections # Add sections

        if not source_article:
            print(f"Warning: 'title' and 'sections' keys not found or empty in original data at index {i}. Skipping entry.")
            continue # Skip if no source article

        if original_abstract is None:
            print(f"Warning: 'abstract' key not found in original data at index {i}. Skipping entry.")
            continue # Skip to the next article if original abstract is missing

        if generated_abstract is None:
            print(f"Warning: 'generated_abstract' key not found in generated data at index {i}. Skipping entry.")
            continue # Skip to the next article if generated abstract is missing


        # Process generated abstract for <think> tag if needed
        if generated_abstract.startswith('<think>'):
            generated_abstract = generated_abstract.split('\n\n')[-1].strip() # Extract after thinking and remove extra whitespace

        src_list.append(source_article) # Append the full article to src_list
        output_list.append(generated_abstract) # Append generated abstract to output_list
        ref_list.append(original_abstract)   # Append original abstract to ref_list

    return src_list, output_list, ref_list


def evaluate_abstracts(original_file_path, generated_file_path):
    """Evaluate abstracts using the provided evaluation framework from two separate files.

    Args:
        original_file_path (str): Path to the JSON file containing original articles and abstracts.
        generated_file_path (str): Path to the JSON file containing generated abstracts.

    Returns:
        dict: Evaluation scores from UniEval.
    """
    # Load the JSON data from two files
    original_data = load_abstracts_from_file(original_file_path)
    generated_data = load_abstracts_from_file(generated_file_path)

    if original_data is None or generated_data is None:
        print("Evaluation aborted due to loading errors.")
        return {} # Return empty dict if loading failed

    # Extract lists for evaluation
    src_list, output_list, ref_list = extract_abstracts(original_data, generated_data)

    if not src_list or not output_list or not ref_list:
        print("No abstracts or source articles extracted for evaluation. Check input files and data structure.")
        return {} # Return empty dict if no abstracts to evaluate

    # Prepare data for pre-trained evaluators
    data = convert_to_json(output_list=output_list,
                          src_list=src_list,
                          ref_list=ref_list)

    # Initialize evaluator for summarization task
    evaluator = get_evaluator('summarization')

    # Get multi-dimensional evaluation scores
    eval_scores = evaluator.evaluate(data, print_result=True)

    return eval_scores


if __name__ == "__main__":
    # Define file paths for original and generated abstracts
    original_article_file = "/content/13_processed_papers.json"  # Replace with your original articles file name
    generated_abstract_file = "/content/drive/Shareddrives/Senior Design Project/13 Paper /Llama3.2_3B/abstracts_llama3.2_3b_20250314_231510.json"   # Replace with your generated abstracts file name

    # Evaluate the abstracts from separate files
    scores = evaluate_abstracts(original_article_file, generated_abstract_file)

    if scores: # Only print if evaluation was successful and scores are returned
        print("\nEvaluation complete!")
    else:
        print("\nEvaluation was not completed successfully. Check error messages.")
```

```python
#GEVAL
import json
import google.generativeai as genai
import time
from typing import List, Dict, Any
import re

# Configure Gemini API
def setup_gemini(api_key: str):
    """Setup Gemini API with the provided key."""
    genai.configure(api_key=api_key)
    model = genai.GenerativeModel('gemini-2.0-flash')
    return model

def load_abstracts(formatted_papers_path: str, mistral_abstracts_path: str) -> List[Dict[str, Any]]:
    """Load and match abstracts from formatted papers and Mistral abstracts."""
    # Load formatted papers
    with open(formatted_papers_path, 'r') as f:
        formatted_papers = json.load(f)

    # Load Mistral abstracts
    with open(mistral_abstracts_path, 'r') as f:
        mistral_abstracts = json.load(f)

    # Create a lookup dictionary for Mistral abstracts
    mistral_dict = {paper['title']: paper['abstract'] for paper in mistral_abstracts}

    # Match papers and create final dataset
    matched_papers = []
    for paper in formatted_papers:
        title = paper['title']
        if title in mistral_dict:
            # Combine sections to create full content
            content = ' '.join(paper['sections'])
            matched_paper = {
                'title': title,
                'content': content,
                'original_abstract': paper['abstract'],
                'generated_abstract': mistral_dict[title]
            }
            matched_papers.append(matched_paper)

    return matched_papers

def create_coherence_prompt(article_content: str, abstract: str, is_original: bool) -> str:
    """Create a prompt for evaluating coherence."""
    abstract_type = "Original" if is_original else "Generated"
    return f"""You are an expert academic reviewer evaluating the coherence of an abstract.

Source Article:
{article_content}

{abstract_type} Abstract:
{abstract}

Evaluation Criteria:
Coherence (1-5) - the collective quality of all sentences. The summary should be well-structured and well-organized, building from sentence to a coherent body of information about a topic.

Evaluation Steps:
1. Read the article carefully and identify the main topic and key points.
2. Read the abstract and compare it to the article.
3. Check if the abstract covers the main topic and key points in a clear and logical order.

Please provide a score between 1-5 and include it in your response like this: "score": X
You can explain your reasoning, but make sure to include the score in the format shown above.

Format your response as JSON:
{{
    "coherence": {{
        "score": X
    }}
}}"""

def create_consistency_prompt(article_content: str, abstract: str, is_original: bool) -> str:
    """Create a prompt for evaluating consistency."""
    abstract_type = "Original" if is_original else "Generated"
    return f"""You are an expert academic reviewer evaluating the consistency of an abstract.

Source Article:
{article_content}

{abstract_type} Abstract:
{abstract}

Evaluation Criteria:
Consistency (1-5) - the factual alignment between the abstract and the source. A factually consistent abstract contains only statements that are entailed by the source document. Penalize abstracts that contain hallucinated facts.

Evaluation Steps:
1. Read the article carefully and identify the main facts and details.
2. Read the abstract and compare it to the article.
3. Check for any factual errors or hallucinated content not supported by the article.

Please provide a score between 1-5 and include it in your response like this: "score": X
You can explain your reasoning, but make sure to include the score in the format shown above."""

def create_fluency_prompt(abstract: str, is_original: bool) -> str:
    """Create a prompt for evaluating fluency."""
    abstract_type = "Original" if is_original else "Generated"
    return f"""You are an expert academic reviewer evaluating the fluency of an abstract.

{abstract_type} Abstract:
{abstract}

Evaluation Criteria:
Fluency (1-3) - the quality of the abstract in terms of grammar, spelling, punctuation, word choice, and sentence structure.

Scoring Scale:
1: Poor - Many errors that make it hard to understand or sound unnatural
2: Fair - Some errors that affect clarity but main points are comprehensible
3: Good - Few or no errors, easy to read and follow

Please provide a score between 1-3 and include it in your response like this: "score": X
You can explain your reasoning, but make sure to include the score in the format shown above."""

def create_relevance_prompt(article_content: str, abstract: str, is_original: bool) -> str:
    """Create a prompt for evaluating relevance."""
    abstract_type = "Original" if is_original else "Generated"
    return f"""You are an expert academic reviewer evaluating the relevance of an abstract.

Source Article:
{article_content}

{abstract_type} Abstract:
{abstract}

Evaluation Criteria:
Relevance (1-5) - selection of important content from the source. The abstract should include only important information from the source document. Penalize abstracts which contain redundancies and excess information.

Evaluation Steps:
1. Read the abstract and source document carefully.
2. Compare the abstract to the source document and identify main points.
3. Assess coverage of main points and presence of irrelevant/redundant information.

Please provide a score between 1-5 and include it in your response like this: "score": X
You can explain your reasoning, but make sure to include the score in the format shown above."""

def evaluate_with_gemini(model, prompt: str) -> Dict[str, Any]:
    """Evaluate using Gemini with a specific prompt and extract score using regex."""
    try:
        response = model.generate_content(prompt)

        # Check if response was successful
        if not response.text:
            print(f"Empty response from Gemini API")
            return None

        # Extract score using regex
        # Look for patterns like "score": X or "score": "X" or "score":X
        score_pattern = r'"score":\s*["\']?(\d+)["\']?'
        match = re.search(score_pattern, response.text)

        if match:
            score = int(match.group(1))
            # Create the appropriate metric dictionary based on the prompt
            if "coherence" in prompt.lower():
                return {"coherence": {"score": score}}
            elif "consistency" in prompt.lower():
                return {"consistency": {"score": score}}
            elif "fluency" in prompt.lower():
                return {"fluency": {"score": score}}
            elif "relevance" in prompt.lower():
                return {"relevance": {"score": score}}
        else:
            print(f"No score found in response: {response.text[:200]}...")
            return None

    except Exception as e:
        print(f"Error in Gemini API call: {str(e)}")
        return None

def evaluate_single_abstract(model, content: str, abstract: str, is_original: bool) -> Dict[str, Any]:
    """Evaluate a single abstract (original or generated) on all metrics."""
    max_retries = 3
    retry_delay = 2  # seconds

    def evaluate_metric(prompt_func, metric_name):
        for attempt in range(max_retries):
            # Special case for fluency which doesn't need content
            if metric_name == "fluency":
                prompt = prompt_func(abstract, is_original)
            else:
                prompt = prompt_func(content, abstract, is_original)

            result = evaluate_with_gemini(model, prompt)
            if result is not None:
                return result
            print(f"Retrying {metric_name} evaluation (attempt {attempt + 1}/{max_retries})...")
            time.sleep(retry_delay)
        return None

    # Evaluate each metric with retries
    coherence_eval = evaluate_metric(create_coherence_prompt, "coherence")
    if not coherence_eval:
        return None

    time.sleep(1)  # Rate limiting
    consistency_eval = evaluate_metric(create_consistency_prompt, "consistency")
    if not consistency_eval:
        return None

    time.sleep(1)  # Rate limiting
    fluency_eval = evaluate_metric(create_fluency_prompt, "fluency")
    if not fluency_eval:
        return None

    time.sleep(1)  # Rate limiting
    relevance_eval = evaluate_metric(create_relevance_prompt, "relevance")
    if not relevance_eval:
        return None

    return {
        'coherence_score': coherence_eval['coherence']['score'],
        'consistency_score': consistency_eval['consistency']['score'],
        'fluency_score': fluency_eval['fluency']['score'],
        'relevance_score': relevance_eval['relevance']['score']
    }

def evaluate_abstracts(formatted_papers_path: str, mistral_abstracts_path: str, api_key: str):
    """Evaluate both original and generated abstracts using Gemini as judge."""
    # Setup Gemini
    model = setup_gemini(api_key)

    # Load the JSON data
    articles = load_abstracts(formatted_papers_path, mistral_abstracts_path)

    # Store results
    evaluations = []

    # Try to load existing checkpoint if it exists
    checkpoint_file = 'evaluation_checkpoint.json'
    try:
        with open(checkpoint_file, 'r') as f:
            checkpoint_data = json.load(f)
            evaluations = checkpoint_data['evaluations']
            start_idx = len(evaluations)
            print(f"Loaded checkpoint with {start_idx} evaluated articles")
    except FileNotFoundError:
        start_idx = 0
        print("Starting fresh evaluation")

    # Process each article
    for i, article in enumerate(articles[start_idx:], start=start_idx):
        print(f"\nEvaluating article {i+1}/{len(articles)}")

        # Get article content and abstracts
        content = article['content']
        original_abstract = article['original_abstract']

        # Clean up generated abstract if it contains thinking process
        generated_abstract = article['generated_abstract']
        if generated_abstract.startswith('<think>'):
            generated_abstract = generated_abstract.split('\n\n')[-1]

        # Evaluate original abstract
        print("Evaluating original abstract...")
        original_scores = evaluate_single_abstract(
            model, content, original_abstract, True
        )

        # Add delay between evaluations
        time.sleep(1)

        # Evaluate generated abstract
        print("Evaluating generated abstract...")
        generated_scores = evaluate_single_abstract(
            model, content, generated_abstract, False
        )

        # Combine results if both evaluations were successful
        if original_scores and generated_scores:
            evaluation = {
                'title': article['title'],
                'original_abstract': {
                    'text': original_abstract,
                    'scores': original_scores
                },
                'generated_abstract': {
                    'text': generated_abstract,
                    'scores': generated_scores
                }
            }
            evaluations.append(evaluation)

            # Save checkpoint after each successful evaluation
            checkpoint_data = {
                'evaluations': evaluations,
                'last_processed_index': i + 1,
                'timestamp': time.strftime('%Y-%m-%d %H:%M:%S')
            }
            with open(checkpoint_file, 'w') as f:
                json.dump(checkpoint_data, f, indent=2)
            print(f"Saved checkpoint after article {i+1}")

        # Add delay before next article
        time.sleep(1)

    # Calculate average scores
    if evaluations:
        avg_scores = {
            'original': {
                'coherence': sum(e['original_abstract']['scores']['coherence_score'] for e in evaluations) / len(evaluations),
                'consistency': sum(e['original_abstract']['scores']['consistency_score'] for e in evaluations) / len(evaluations),
                'fluency': sum(e['original_abstract']['scores']['fluency_score'] for e in evaluations) / len(evaluations),
                'relevance': sum(e['original_abstract']['scores']['relevance_score'] for e in evaluations) / len(evaluations)
            },
            'generated': {
                'coherence': sum(e['generated_abstract']['scores']['coherence_score'] for e in evaluations) / len(evaluations),
                'consistency': sum(e['generated_abstract']['scores']['consistency_score'] for e in evaluations) / len(evaluations),
                'fluency': sum(e['generated_abstract']['scores']['fluency_score'] for e in evaluations) / len(evaluations),
                'relevance': sum(e['generated_abstract']['scores']['relevance_score'] for e in evaluations) / len(evaluations)
            }
        }

        # Save final results
        final_results = {
            'individual_evaluations': evaluations,
            'average_scores': avg_scores
        }
        with open('geval_comparison_results.json', 'w') as f:
            json.dump(final_results, f, indent=2)

        return evaluations, avg_scores
    else:
        print("No successful evaluations completed")
        return [], {}

if __name__ == "__main__":
    # Replace with your Gemini API key
    API_KEY = ""

    # Evaluate the abstracts
    formatted_papers_file = "formatted_papersv2.json"
    mistral_abstracts_file = "Mistral_large_abstracts.json"
    evaluations, avg_scores = evaluate_abstracts(formatted_papers_file, mistral_abstracts_file, API_KEY)

    # Print summary
    print("\nEvaluation Summary:")
    print("==================")
    print("\nOriginal Abstracts:")
    for metric, score in avg_scores['original'].items():
        print(f"{metric.capitalize()}: {score:.2f}")

    print("\nGenerated Abstracts:")
    for metric, score in avg_scores['generated'].items():
        print(f"{metric.capitalize()}: {score:.2f}")

    print(f"\nDetailed results saved to geval_comparison_results.json")
```
